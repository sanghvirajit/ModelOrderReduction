```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('mnist_train.csv')
train = pd.DataFrame(train)

test = pd.read_csv('mnist_test.csv')
test = pd.DataFrame(test)

x_train = pd.read_csv('x_train.csv')
X_train = pd.DataFrame(x_train)

y_train = pd.read_csv('y_train.csv')
Y_train = pd.DataFrame(y_train)

X_train = X_train.values
Y_train = Y_train.values
```

```python
plt.imshow(X_train[0].reshape(28, 28), cmap='Greys')
plt.show()
```

```python
images_6 = X_train[36017:41934]
```

```python
images_6.shape
```

```python
for i in range(10):
    plt.imshow(images_6[i].reshape(28, 28), cmap='Greys')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(images_6)
print(standardized_data.shape)
```

```python
sample_data_images = standardized_data
sample_data_images_6 = sample_data_images.T
sample_data_images_6.shape
```

```python
co_variance_matrix_6 = np.dot(sample_data_images_6.T, sample_data_images_6)
co_variance_matrix_6.shape
```

```python
from scipy.linalg import eigh

eigen_values, eigen_vectors = eigh(co_variance_matrix_6)
print("Shape of eigen vectors = ", eigen_vectors.shape)
```

```python
percentage_var_explained = eigen_values / np.sum(eigen_values)
cum_var_explained = np.cumsum(percentage_var_explained)

plt.plot(cum_var_explained, linewidth=2)
plt.xlabel('principle no of images')
plt.ylabel('variance')
plt.grid()
plt.show()
```

```python
images_0 = X_train[0:5922]
```

```python
for i in range(10):
    plt.imshow(images_0[i].reshape(28, 28), cmap='Greys')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data_0 = StandardScaler().fit_transform(images_0)
standardized_data_0.shape
```

```python
sample_data_0 = standardized_data_0
co_variance_matrix_0 = np.matmul(sample_data_0, sample_data_0.T)
co_variance_matrix_0.shape
```

```python
from scipy.linalg import eigh
eigen_values, eigen_vectors = eigh(co_variance_matrix_0)
```

```python
variance_percentage = eigen_values / np.sum(eigen_values)
variance_percentage_cum = np.cumsum(variance_percentage)
```

```python
plt.plot(variance_percentage_cum)
plt.xlabel('Images of shape 0')
plt.ylabel('variance')
plt.grid()
plt.show()
```
