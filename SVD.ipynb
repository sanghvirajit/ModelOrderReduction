```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import svd
```

```python
train = pd.read_csv('mnist_train.csv')
test = pd.read_csv('mnist_test.csv')
```

```python
x_train = train.drop(columns='label')
x_test = test.drop(columns='label')
```

```python
y_train = train['label']
y_test = test['label']
```

```python
x_train = x_train.values
x_test = x_test.values
```

```python
U,S,V = svd(x_train[0].reshape(28, 28))
```

```python
n = len(x_train[0].reshape(28, 28))
imgs_1 = []
for i in range(n):
    imgs_1.append(S[i]*np.outer(U[:,i],V[i]))
    print(i)
    plt.imshow(imgs_1[i], cmap='Greys')
    plt.show()
```

```python
n = len(x_train[0].reshape(28, 28))
combined_imgs_1 = []
for i in range(n):
    img = sum(imgs_1[:i+1])
    combined_imgs_1.append(img)
    print(i)
    plt.imshow(combined_imgs_1[i], cmap='Greys')
    plt.show()
```
